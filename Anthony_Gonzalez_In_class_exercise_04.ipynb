## 2.1 Write a Python program to remove leading zeros from an IP address. (4 points)

ip = "260.08.094.109"

```python
ip = "260.08.094.109" #Creates and ip address string
new_ip = [] #Empty list for to hold new ip address

x = ip.split(".") #Splits the ip address at the '.' and creates a list

#For loop iterates through the list, converts each string into an in removing leading zeros, and adds to empty list
for i in x:
  new_ip.append(str(int(i)))

".".join(new_ip) #joins the new_ip list with a '.' delimiter
```

## 2.2 Write a Python Program to extract all the years from the following sentence. (4 points)

sentence = "The 2010s were a dramatic decade, filled with ups and downs, more than 1000 stroies have happened. As the decade comes to a close, Insider took a look back at some of the biggest headline-grabbing stories, from 2010 to 2019. The result was 119 news stories that ranged from the heartwarming rescue of a Thai boys' soccer team from a flooded cave to the divisive election of President Donald Trump."

```python
import re #imports RegEx module to use finall function

#Sets string from example provided
sentence = "The 2010s were a dramatic decade, filled with ups and downs, more than 1000 stroies have happened. As the decade comes to a close, Insider took a look back at some of the biggest headline-grabbing stories, from 2010 to 2019. The result was 119 news stories that ranged from the heartwarming rescue of a Thai boys' soccer team from a flooded cave to the divisive election of President Donald Trump."

#Uses findall function from RegEx module to find four digit numbers from string that begin with 2 and all other digts can range from 0-9
years = re.findall('[2][0-9][0-9][0-9]', sentence)

print(years)
```
